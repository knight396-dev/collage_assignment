1) Practise Pandas Series Create a Pandas Series from Dictionary Create a Pandas Series from Lists Access the elements of a Series in Pandas

```python
import pandas as pd

print("\n1. Pandas Series from Dictionary")
data_dict = {
    "name": "Shourya",
    "age": 21,
    "course": "B.Tech"
}
series_dict = pd.Series(data_dict)
print(series_dict)

print("\n2. Pandas Series from List")
data_list = [10, 20, 30, 40, 50]
series_list = pd.Series(data_list)
print(series_list)

print("\n3. Access Elements of Series")
print("First element:", series_list[0])
print("Element using iloc:", series_list.iloc[2])
print("Element from dictionary Series:", series_dict["name"])
```

2) DataFrames Make a Pandas DataFrame with a two-dimensional Python list Create DataFrame from Python dict Create Pandas dataframe using list of lists Create a Pandas dataframe using list of tuples Create a Pandas DataFrame from List of Dicts

```python
print("\n4. DataFrame using Two-Dimensional Python List")
data_2d = [
    ["Shourya", 21],
    ["Aman", 22],
    ["Rahul", 20]
]
df_2d = pd.DataFrame(data_2d, columns=["Name", "Age"])
print(df_2d)

print("\n5. DataFrame from Python Dictionary")
data_dict_df = {
    "Name": ["Shourya", "Aman", "Rahul"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}
df_dict = pd.DataFrame(data_dict_df)
print(df_dict)

print("\n6. DataFrame using List of Lists")
list_of_lists = [
    [1, "Python", 95],
    [2, "Pandas", 90],
    [3, "NumPy", 88]
]
df_list_lists = pd.DataFrame(list_of_lists, columns=["ID", "Subject", "Score"])
print(df_list_lists)

print("\n7. DataFrame using List of Tuples")
list_of_tuples = [
    (1, "Shourya", "CSE"),
    (2, "Aman", "IT"),
    (3, "Rahul", "AI")
]
df_tuples = pd.DataFrame(list_of_tuples, columns=["Roll No", "Name", "Branch"])
print(df_tuples)

print("\n8. DataFrame from List of Dictionaries")
list_of_dicts = [
    {"Name": "Shourya", "Age": 21, "City": "Jaipur"},
    {"Name": "Aman", "Age": 22, "City": "Delhi"},
    {"Name": "Rahul", "Age": 20, "City": "Mumbai"}
]
df_list_dicts = pd.DataFrame(list_of_dicts)
print(df_list_dicts)
```

3)Data iteration Different ways to iterate over rows in Pandas Dataframe Selecting rows in pandas DataFrame based on conditions Select any row from a Dataframe using iloc[] Limited rows selection with given column Drop rows from the dataframe based on certain condition applied on a column Insert row at given position in Pandas Dataframe Create a list from rows in Pandas dataframe

```python
df = pd.DataFrame({
    "Name": ["Shourya", "Aman", "Rahul", "Riya", "Neha"],
    "Age": [21, 22, 20, 23, 19],
    "Marks": [85, 90, 78, 92, 70],
    "City": ["Jaipur", "Delhi", "Mumbai", "Pune", "Jaipur"]
})

print("\nOriginal DataFrame")
print(df)

print("\n9. Iterate over rows using iterrows()")
for index, row in df.iterrows():
    print(index, row["Name"], row["Marks"])

print("\n10. Iterate over rows using itertuples()")
for row in df.itertuples():
    print(row.Name, row.Age, row.Marks)

print("\n11. Selecting Rows Based on Condition")
marks_above_80 = df[df["Marks"] > 80]
print(marks_above_80)

print("\n12. Select Any Row using iloc[]")
print(df.iloc[2])

print("\n13. Limited Rows Selection with Given Column")
limited_selection = df.loc[0:2, ["Name", "Marks"]]
print(limited_selection)

print("\n14. Drop Rows Based on Condition")
df_dropped = df[df["Marks"] >= 80]
print(df_dropped)

print("\n15. Insert Row at Given Position")
new_row = pd.DataFrame([{
    "Name": "Karan",
    "Age": 24,
    "Marks": 88,
    "City": "Kota"
}])

position = 2

df_inserted = pd.concat(
    [df.iloc[:position], new_row, df.iloc[position:]],
    ignore_index=True
)

print(df_inserted)

print("\n16. Create List from Rows in DataFrame")
rows_list = df.values.tolist()
print(rows_list)
```
