3) Replace NaN with column average

```python
import numpy as np

arr = np.array([
    [1, 2, np.nan],
    [4, np.nan, 6],
    [7, 8, 9]
])

col_mean = np.nanmean(arr, axis=0)

# replace NaN column-wise
inds = np.where(np.isnan(arr))

arr[inds] = np.take(col_mean, inds[1])

print(arr)
```

5) Mean, Median, Mode of two 2D arrays

```python
import numpy as np

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

combined = np.concatenate((arr1, arr2))

# Mean
mean_val = np.mean(combined)

# Median
median_val = np.median(combined)

# Mode (NumPy-only workaround)
values, counts = np.unique(combined, return_counts=True)
mode_val = values[np.argmax(counts)]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

6) Solve equations using linalg (Matrix method)

x - 2y + 3z = 9
-x + 3y - z = -6
2x - 5y + 5z = 17

```python
import numpy as np

A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([9, -6, 17])

# Using inverse method
X = np.linalg.inv(A).dot(B)

print("Solution (x,y,z):", X)

#X = np.linalg.solve(A, B)
```
