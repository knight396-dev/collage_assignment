```python
t = (1, 2, 3)
result = t * 3 
print("Repeated Tuple:", result)
```

```python
t1 = (1, 2) 
t2 = (3, 4)
t3 = (5, 6) 
result = t1 + t2 + t3 
print("Combined Tuple:", result)
```

```python
t = (10, 20, 30, 40) 
element = 30
if element in t:
    print(element, "exists in the tuple.") 
else:
    print(element, "does not exist in the tuple.")
```

```python
t = (12, 8, 25, 4, 18) 

total = 0 
highest = t[0]
lowest = t[0] 
for i in t:
    total += i
    if i > highest: 
        highest = i 
    if i < lowest: 
        lowest = i 
        
print("Total:", total)
print("Highest:", highest) 
print("Lowest:", lowest)
```

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for i in n:
    if i > 10:
        result += (i,)

print(result)
```

```python
s = {"cat", "dog", "bird", "fish"}

count = 0

for i in s:
    count += 1

print("Number of elements:", count)
```
