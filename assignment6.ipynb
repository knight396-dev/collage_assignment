```python
#game : pixel drawing using turtle module

'''


recomenddation   : run it on python idle insted od jupyter notebook because it is not working on jupyter notebook
 for better experince run it on python idle insted od jupyter notebook because it is not working on jupyter notebook

'''


import turtle

screen = turtle.Screen() # to greamt a game window 

#to set size ofthe window
screen.setup(width=800, height=600)


# player setup 
player = turtle.Turtle()
player.shape("turtle")
player.color("blue")
#player.penup()  # donesent draw the line 
player. goto(0, -250) # initalization

def move_left():
    x = player.xcor() # gets current position
    x = x - 20
    player.setx(x) # change posrtion 

def move_right():
    x = player.xcor()
    x = x + 20
    player.setx(x)

def move_up():
    y = player.ycor() # gets current position
    y = y + 20
    player.sety(y) # change posrtion 

def move_down():
    y = player.ycor()
    y = y - 20
    player.sety(y)
    
    
is_drawing = False
player.penup()

def toggle_pen():
    global is_drawing
    
    if is_drawing == False:
        player.pendown()
        is_drawing = True
        print("drawing on")
    else:
        player.penup()
        is_drawing = False
        print("drawing off")


# prom;lem two : turtuel usle Left insted of left
screen.listen()
screen.onkeypress(move_left,"Left")
screen.onkeypress(move_right,"Right")
screen.onkeypress(move_up,"Up")
screen.onkeypress(move_down,"Down")

screen.onkeypress(toggle_pen,"d")


screen.mainloop() # NOTHING RUNS AFTER THAT

```

```python
import requests

print("-------free api data fetch-----------")

# 1 . official joke api

joke_url = "https://official-joke-api.appspot.com/random_joke"
joke_response = requests.get(joke_url)
joke_data = joke_response.json()
print("\n ==== joke api data====")
print(f"setup : {joke_data['setup']}")
print(f"punchline : {joke_data['punchline']}")


# dad jokes api

dad_joke_url = "https://icanhazdadjoke.com/"
headers = {"Accept": "application/json"}
dad_joke_response = requests.get(dad_joke_url, headers=headers)
dad_joke_data = dad_joke_response.json()
print("\n ==== dad joke api data====")
print(f"joke : {dad_joke_data['joke']}")

# 3. JSONplaceholder api

post_url = "https://jsonplaceholder.typicode.com/posts/1"
post_response = requests.get(post_url)  
post_data = post_response.json()

print("\n ==== JSONplaceholder api data====")
print(f"userId : {post_data['userId']}")
print(f"id : {post_data['id']}")
print(f"title : {post_data['title']}")
print(f"body : {post_data['body']}")

```
