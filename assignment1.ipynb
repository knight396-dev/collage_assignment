```python
'''
1) Write a python program that takes in a student name, class. It should also take in five subject marks of the students and find the total mark and percentage. Display a result in such a way that their name, class,  and percentage are printed.

'''
```

```python
print("====welcome to the system====")

#details input
name = input("enter the name: ")
Class = int(input("enter the year: "))

# subject marks input
marks = []
for i in range(5):
    mark = int(input(f"enter the mark of subject {i+1}: "))
    marks.append(mark)
# total and percentage calculation
total = sum(marks) 
percentage = (total / 500) * 100

# result display
print(f"Name: {name}")
print(f"Class: {Class}")
print(f"Percentage: {percentage:.2f}%")
```

```python
'''
 Input 2 strings concatinate them and store in another variable. then perform generally used string methods on it like

lower(), upper(), title(), swapcase(), capitalize(), casefold(), center(), count(), endswith(), find(), isalnum(), isdigit(), isnumeric(), isspace(), replace()
'''
```

```python
# input two strings
str1 = input("enter the first string: ")
str2 = input("enter the second string: ")

# concatenate the strings
concatenated_str = str1 + " " + str2
# perform string methods
print("Concatenated String:", concatenated_str)
```

```python
print("Lowercase:", concatenated_str.lower())
print("Uppercase:", concatenated_str.upper())
print("Title Case:", concatenated_str.title())
print("Swap Case:", concatenated_str.swapcase())
print("Capitalized:", concatenated_str.capitalize())
print("Casefolded:", concatenated_str.casefold())
print("Centered:", concatenated_str.center(50))
print("Count of 'a':", concatenated_str.count('a'))
print("Ends with 'a':", concatenated_str.endswith('a'))
print("Find 'a':", concatenated_str.find('a'))
print("Is Alphanumeric:", concatenated_str.isalnum())
print("Is Digit:", concatenated_str.isdigit())
print("Is Numeric:", concatenated_str.isnumeric())
print("Is Space:", concatenated_str.isspace())
print("Replace 'a' with 'x':", concatenated_str.replace('a', 'x'))
```
