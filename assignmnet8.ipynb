 2)Create two DataFrames, df1 and df2, with a common column (e.g., 'ID'). >Perform an inner merge on this common column and display the resulting DataFrame. >Perform a left join of df1 and df2 on the 'ID' column. Explain how missing values are handled in the resulting DataFrame. Right Join and Index-Based Join. >Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys.

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Aman", "Riya", "Shourya", "Neha"]
})

df2 = pd.DataFrame({
    "ID": [3, 4, 5, 6],
    "Marks": [85, 90, 75, 88]
})

print("\nDataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)


# Inner Merge
inner_merge = pd.merge(df1, df2, on="ID", how="inner")
print("\nInner Merge:")
print(inner_merge)


# Left Join
left_join = pd.merge(df1, df2, on="ID", how="left")
print("\nLeft Join:")
print(left_join)

print("""
In left join, all rows from the left DataFrame are kept.
If matching values are not found in the right DataFrame,
Pandas fills the missing values with NaN.
""")


# Right Join
right_join = pd.merge(df1, df2, on="ID", how="right")
print("\nRight Join:")
print(right_join)


# Index-Based Join
df1_index = df1.set_index("ID")
df2_index = df2.set_index("ID")

index_join = df1_index.join(df2_index, how="inner")
print("\nIndex-Based Join:")
print(index_join)

# Merging with Multiple Keys
students = pd.DataFrame({
    "ID": [1, 1, 2, 2],
    "Subject": ["Python", "DBMS", "Python", "DBMS"],
    "Name": ["Aman", "Aman", "Riya", "Riya"]
})

marks = pd.DataFrame({
    "ID": [1, 1, 2, 2],
    "Subject": ["Python", "DBMS", "Python", "DBMS"],
    "Marks": [88, 75, 92, 80]
})

multiple_key_merge = pd.merge(
    students,
    marks,
    on=["ID", "Subject"],
    how="inner"
)

print("\nMerge with Multiple Keys:")
print(multiple_key_merge)
```

3) Create three DataFrames. Vertically concatenate two of them using pd.concat(), then merge the resulting DataFrame with the third DataFrame on a common key. T Understand join() vs. merge(). Also Explain the primary differences between df.join() and pd.merge(),

```python
# 3) Concatenation and Merge
df_a = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["Aman", "Riya"]
})

df_b = pd.DataFrame({
    "ID": [3, 4],
    "Name": ["Shourya", "Neha"]
})

df_c = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Course": ["Python", "Pandas", "NumPy", "Data Science"]
})

vertical_concat = pd.concat([df_a, df_b], ignore_index=True)

print("\nVertical Concatenation:")
print(vertical_concat)

final_merge = pd.merge(vertical_concat, df_c, on="ID", how="inner")

print("\nFinal Merged DataFrame:")
print(final_merge)
```
